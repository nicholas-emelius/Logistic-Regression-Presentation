```python
# Step 1: Import the required modules

from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

```python
# Step 2: Generate the dataset

X, y = make_blobs(n_samples=125, centers=2, cluster_std=0.60, random_state=0)
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=20, random_state=0)
```

```python
# Step 3: Visualize the data

plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y, cmap='rainbow')
```

```python
# Step 4: Create SVC

svc = SVC(kernel='linear')
svc.fit(train_X, train_y)
```

```python
# Step 5: PLot the hyperplane

plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y, cmap='rainbow')
ax = plt.gca()
xlim = ax.get_xlim()
ax.scatter(test_X[:, 0], test_X[:, 1], c=test_y, cmap='rainbow', marker='s')

# Step 6: Plot the hyperplane

w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(xlim[0], xlim[1])
yy = a * xx - (svc.intercept_[0] / w[1])

b = svc.support_vectors_[0]
yy_down = a * xx +(b[1] - a * b[0])
c = svc.support_vectors_[-1]
yy_up = a * xx + (c[1] - a * c[0])
plt.plot(xx,yy)
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
plt.show
```

```python
# Step 6: Make prediction using the model

pred_y = svc.predict(test_X)
```

```python
# Step 7: Display the confusion matrix

confusion_matrix(test_y, pred_y)
```
